```python
#Generate a list of all 49 companies
companies = data['Company'].unique()
```

```python
#Aggregate the dataset over each company by generating mean, stdDev, Min and Max values for each company
#Save aggregared data to a file

agg_data = pd.DataFrame(columns=['mean','std','min','max', 'Company'])
i = 0
for company in companies:
    small_data = data[data['Company'] == company]
    desc = (small_data.describe()['Adj. Close'])
    agg_data.loc[i] = [desc[1], desc[2], desc[3], desc[7], company]
    i+=1

print (agg_data)
agg_data.to_csv('Aggregated_AdjClose.csv')

```

```python
train_data = agg_data[["mean", "std", "min", "max"]]
train_data.head()
```

```python
#Set up K-Means and iterate from 1-20 clusters to decide which is the best number of clusters
sse = []
ks = []
for k in range(1,20,1):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(train_data)
    sse.append(kmeans.inertia_)
    ks.append(k)
```

```python
#Plot the SSE graph
ax = sb.lineplot(x = ks, y = sse)
ax.set(xlabel = 'K clusters', ylabel = 'SSE')
plt.show()
```

```python
#Run clustering on the data and add the values of the clusters to original data
kmeans = KMeans(n_clusters=4)
kmeans.fit(train_data)
values = kmeans.predict(train_data)
agg_data['Clusters'] = values
```
